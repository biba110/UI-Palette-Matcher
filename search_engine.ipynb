```python
import pandas as pd
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer

# 1. Load the dataset you just generated
print("Loading dataset...")
df = pd.read_csv("ui_dataset_batched.csv")
moods = df["mood_industry"].tolist()

# 2. Load the AI language model (this downloads a small model the first time)
print("Loading SentenceTransformer model...")
model = SentenceTransformer('all-MiniLM-L6-v2')

# 3. Convert text to embeddings (numbers)
print("Converting text to vectors...")
embeddings = model.encode(moods, convert_to_numpy=True)

# 4. Set up the FAISS vector database
print("Building FAISS index...")
dimension = embeddings.shape[1] 
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

# 5. Create the search function
def search_palettes(user_query, top_k=3):
    query_vector = model.encode([user_query], convert_to_numpy=True)
    distances, indices = index.search(query_vector, top_k)
    
    print(f"\n--- Top {top_k} Results for: '{user_query}' ---")
    for i in range(top_k):
        row_index = indices[0][i]
        match = df.iloc[row_index]
        print(f"\nMatch {i+1}: {match['mood_industry']}")
        print(f"Colors: Primary {match['primary_color']} | Background {match['background_color']}")
        print(f"Fonts: {match['heading_font']} / {match['body_font']}")

# 6. Test it out!
search_palettes("A clean and professional dental clinic")
```
